```python
corr = errors.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
from scipy. stats import spearmanr
```

```python
corr_list = [(corr.iloc[row,col]) for row in range(1, corr.count()[0]) \
             for col in range(row)]
max3 = sorted(corr_list, key=lambda x: abs(x), reverse=True)[:10]
headings = [] 
for col, data in corr.items():
  headings.append(col)
print('Top 10 correlations:', '\n')
corr_dict = {}
for col, data in corr.items(): 
  for idx, el in enumerate(data):
    if el in max3 and (headings[idx], col) not in corr_dict.keys() \
    and (col, headings[idx]) not in corr_dict.keys():
          corr_dict[headings[idx], col] = el
for key, value in corr_dict.items():
  print(' and '.join(key), ': ', value)
  rho, p = spearmanr(errors[key[0]], errors[key[1]])
  print('According to Spearman:', rho)
  print('p =', p) 
  if p < 0.5:
    print('Корреляция значима')
  print()
```
